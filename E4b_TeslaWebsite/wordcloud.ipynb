```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# Read text for wordcloud
with open('wordCloud.txt', 'r') as f:
    text = f.read()
```

```python
# Wordcloud settings
wordcloud = WordCloud(width=2000, height=1000, random_state=4, max_words=50, background_color="white")
# Process words and generate wordcloud
wc_dict = wordcloud.process_text(text)
wordcloud = wordcloud.generate_from_frequencies(wc_dict)

# Plot figure and save
plt.figure(figsize=(30, 16))
plt.imshow(wordcloud)
plt.axis("off")
plt.savefig("wordcloud" + ".png", bbox_inches='tight')
```
